3.1. Chuyển định dạng của cột Date thành datetime

```python
from datetime import datetime
import pandas as pd

# Danh sách các định dạng ngày hợp lệ
date_formats = [
    '%d/%m/%Y %H:%M',
    '%d/%m/%Y %I:%M:%S %p',
    '%d-%m-%Y %H:%M:%S',
    '%Y-%m-%d %H:%M:%S',
    '%d/%m/%Y',
    '%d-%m-%Y',
    '%m/%d/%Y %I:%M:%S %p'
]

# Hàm chuyển chuỗi thành datetime nếu khớp
def try_parse_date(value):
    value = str(value).strip()
    for fmt in date_formats:
        try:
            return datetime.strptime(value, fmt)
        except:
            continue
    return value  # Không phải ngày thì giữ nguyên

# Áp dụng to_datetime cho từng cột
for col in df.columns:
    try:
        df[col] = pd.to_datetime(df[col], errors='ignore')
    except:
        pass


```

3.2. Dời vị trí các giá trị cho đúng cột

```python
import pandas as pd
import re

# Danh sách cột cần kiểm tra (bao gồm 'Date' như một cột có thể sai)
columns_to_check = ['MA URL', 'MA Referrer', 'ma_path', 'IP Address', 'cuid', 'Message Id', 'Date']

# Định nghĩa pattern regex đúng cho từng cột
patterns = {
    'MA URL': r'^https?://',
    'MA Referrer': r'utm_|^https?://|^[\w\d.-]+$',
    'ma_path': r'^/.*?/$|^utm',
    'IP Address': r'^\d{1,3}(\.\d{1,3}){3}$',
    'cuid': r'^[a-f0-9]{8}-[a-f0-9]{4}-[1-5][a-f0-9]{3}-[89ab][a-f0-9]{3}-[a-f0-9]{12}$',
    'Message Id': r'^<[\w\d.@]+>$',
    'Date': r'.*'  # Tạm thời để .* (sẽ dùng pd.to_datetime thay thế)
}

# Hàm kiểm tra regex
def match_pattern(val, pattern):
    if pd.isna(val) or str(val).strip() == '':
        return False
    return bool(re.match(pattern, str(val).strip()))

# Hàm kiểm tra giá trị có thể là ngày
def is_valid_datetime(val):
    try:
        pd.to_datetime(val, errors='raise', infer_datetime_format=True)
        return True
    except:
        return False

# Hàm xử lý từng dòng
def clean_row(row):
    # Duyệt tất cả các cột, kể cả 'Date'
    for col in columns_to_check:
        val = str(row[col]).strip()
        if val == '':
            continue

        # Trường hợp đặc biệt MA URL
        if col == 'MA URL':
            if not match_pattern(val, patterns['MA URL']):
                row[col] = ''
            continue

        # Nếu cột là Date, kiểm tra datetime hợp lệ
        if col == 'Date':
            if is_valid_datetime(val):
                continue  # giữ lại
            # Nếu không phải ngày, thử phân loại lại

        # Nếu hợp với pattern hiện tại → đúng rồi
        if col != 'Date' and match_pattern(val, patterns[col]):
            continue

        # Nếu sai, thử xem hợp cột khác không
        for tgt_col in columns_to_check:
            if tgt_col == col:
                continue
            if tgt_col == 'Date':
                if is_valid_datetime(val):
                    if pd.isna(row['Date']) or row['Date'] == '':
                        row['Date'] = pd.to_datetime(val, errors='coerce', infer_datetime_format=True)
                        row[col] = ''
                        break
            elif match_pattern(val, patterns[tgt_col]):
                if pd.isna(row[tgt_col]) or row[tgt_col] == '':
                    row[tgt_col] = val
                    row[col] = ''
                    break

    return row

# Áp dụng
df = df.apply(clean_row, axis=1)

# Đảm bảo cột Date là datetime
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
```

Đổi thứ tự của cột Date, đang bị sai

```python
import pandas as pd
from pandas import Timestamp

# Sao lưu cột gốc để xử lý an toàn (tùy chọn)
df['Date_original'] = df['Date']

# Bước 1: Parse tạm thành datetime
parsed_dates = pd.to_datetime(df['Date'], format='%d-%m-%Y', errors='coerce')

# Bước 2: Tạo phiên bản mới đã hoán đổi ngày và tháng nếu hợp lệ
def swap_day_month(x):
    try:
        return Timestamp(year=x.year, month=x.day, day=x.month)
    except:
        return pd.NaT

swapped = parsed_dates.apply(swap_day_month)

# Bước 3: Format lại chuỗi, chỉ áp dụng nếu hoán đổi thành công, còn lại giữ nguyên
df['Date'] = swapped.dt.strftime('%d-%m-%Y').combine_first(df['Date_original'])
```
